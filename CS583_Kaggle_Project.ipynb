### 1.3 Data Preprocessing

```python
def normalize_data(train, test):
    '''
        Normalizing the data using the Mean - Standard Deviation Method
        x = x - mean / std
    '''
    mean = train.iloc[:,2:].mean(axis=0)
    std = train.iloc[:,2:].std(axis=0)
    
    train = train.iloc[:,2:] - mean
    train = train.iloc[:,2:] / std
    
    return train
#     test -= mean
#     test /= std
```

## 2. Building The Model
-------
### 2.1 Define the model

```python
def build_model(train_dim, layer_size=200):
    model = Sequential()
    model.add(Dense(layer_size,input_dim=200, kernel_initializer='normal'))
    model.add(Activation('relu'))
    model.add(Dropout(0.3))
    
    model.add(Dense(layer_size))
    model.add(BatchNormalization())
    model.add(Activation('relu'))
    model.add(Dropout(.3))
    
    
    model.add(Dense(layer_size))
    model.add(BatchNormalization())
    model.add(Activation('relu'))
    model.add(Dropout(0.3))
    
#     model.add(Dense(512))
#     model.add(BatchNormalization())
#     model.add(Activation('relu'))
#     model.add(Dropout(0.4))
    
    model.add(Dense(1, activation='sigmoid'))
    model.compile(optimizer= optimizers.Adamax(lr=1E-4), loss='binary_crossentropy', metrics=['acc'])
    
    return model    
```

```python
model = build_model(200)
model.summary()
```

### 2.4 Improving The Baseline With K-Fold Method

```python
del model
```

```python
k = 4
folds = list(KFold(n_splits=k, shuffle=True, random_state=1).split(df_train.iloc[:,2:202], df_train.iloc[:,1]))
```

```python
for k , (train_index, valid_index) in enumerate(folds):
    print('Fold # %d' % k)
    x_train = df_train.iloc[train_index,2:202]
    y_train = df_train.iloc[train_index,1]
    
    x_valid = df_train.iloc[valid_index,2:202]
    y_valid = df_train.iloc[valid_index,1]
    
    model = build_model(200)
    
    history = model.fit(x_train, y_train, epochs=30, batch_size=64, validation_data=(x_valid,y_valid), verbose=1)
    
    acc = history.history['acc']
    val_acc = history.history['val_acc']
    
    epochs = range(len(acc))
    
    plt.plot(epochs ,acc , 'bo', label='Training acc')
    plt.plot(epochs, val_acc, 'r', label='Validation acc')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.legend()
    plt.show()
    
    print(model.evaluate(x_valid, y_valid))
    
```
